```python
pdf_folder = "/content/drive/MyDrive/Penalaran Komputer 202310370311281/tugas subcpmk3/PenalaranKomputer_CBR/01_Dataset_PDF"

txt_folder = "/content/drive/MyDrive/Penalaran Komputer 202310370311281/tugas subcpmk3/PenalaranKomputer_CBR/02_Dataset_Text"
```

```python
pdf_files = sorted([
    f for f in os.listdir(pdf_folder)
    if f.endswith(".pdf")
])

print("Jumlah PDF:", len(pdf_files))
```

```python
pdf_files[:5]
```

```python
import pdfplumber
import os

for pdf_file in pdf_files:

    pdf_path = os.path.join(pdf_folder, pdf_file)

    all_text = ""

    try:

        with pdfplumber.open(pdf_path) as pdf:

            for page in pdf.pages:

                page_text = page.extract_text()

                if page_text:
                    all_text += page_text + "\n"

        txt_name = pdf_file.replace(".pdf", ".txt")

        txt_path = os.path.join(txt_folder, txt_name)

        with open(txt_path, "w", encoding="utf-8") as f:
            f.write(all_text)

        print(f"Berhasil: {txt_name}")

    except Exception as e:

        print(f"Gagal: {pdf_file}")
        print(e)
```
