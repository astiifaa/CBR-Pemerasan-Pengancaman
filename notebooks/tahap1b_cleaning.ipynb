```python
import os

input_folder = "/content/drive/MyDrive/Penalaran Komputer 202310370311281/tugas subcpmk3/PenalaranKomputer_CBR/02_Dataset_Text"

output_folder = "/content/drive/MyDrive/Penalaran Komputer 202310370311281/tugas subcpmk3/PenalaranKomputer_CBR/02_Dataset_Text_Clean"

os.makedirs(output_folder, exist_ok=True)
```

```python
import re
```

```python
for file_name in os.listdir(input_folder):

    if not file_name.endswith(".txt"):
        continue

    input_path = os.path.join(input_folder, file_name)

    with open(input_path, "r", encoding="utf-8") as f:
        text = f.read()

    # hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text)

    # hapus halaman
    text = re.sub(r'Halaman \d+ dari \d+ halaman', ' ', text)

    # hapus disclaimer
    text = re.sub(
        r'Disclaimer.*?Halaman \d+',
        ' ',
        text,
        flags=re.IGNORECASE
    )

    # hapus url
    text = re.sub(
        r'putusan\.mahkamahagung\.go\.id',
        ' ',
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(r'\s+', ' ', text)

    output_path = os.path.join(output_folder, file_name)

    with open(output_path, "w", encoding="utf-8") as f:
        f.write(text)

print("Cleaning selesai")
```
