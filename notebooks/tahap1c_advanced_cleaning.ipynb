```python
import os
import re
```

```python
INPUT_FOLDER = "/content/drive/MyDrive/Penalaran Komputer 202310370311281/tugas subcpmk3/PenalaranKomputer_CBR/02_Dataset_Text_Clean"

OUTPUT_FOLDER = "/content/drive/MyDrive/Penalaran Komputer 202310370311281/tugas subcpmk3/PenalaranKomputer_CBR/02_Dataset_Text_Final"

os.makedirs(
    OUTPUT_FOLDER,
    exist_ok=True
)
```

```python
def advanced_clean(text):

    text = str(text)

    # hapus pola huruf satuan
    text = re.sub(
        r'(?:\b[a-zA-Z]\b\s*){10,}',
        ' ',
        text
    )

    # hapus watermark
    patterns = [

        r'Direktori Putusan Mahkamah Agung Republik Indonesia',

        r'putusan\.mahkamahagung\.go\.id',

        r'Disclaimer',

        r'Halaman\s+\d+\s+dari\s+\d+',

        r'MAHKAMAH AGUNG R\.I\.',

        r'Untuk Salinan',

        r'Panitera Pengganti',

        r'PANITERA MUDA PIDANA',

        r'NIP\.\s*\d+'
    ]

    for p in patterns:
        text = re.sub(
            p,
            ' ',
            text,
            flags=re.IGNORECASE
        )

    # rapikan spasi
    text = re.sub(
        r'\s+',
        ' ',
        text
    )

    return text.strip()
```

```python
files = sorted(
    os.listdir(INPUT_FOLDER)
)

for file in files:

    if file.endswith(".txt"):

        path = os.path.join(
            INPUT_FOLDER,
            file
        )

        with open(
            path,
            "r",
            encoding="utf-8",
            errors="ignore"
        ) as f:

            text = f.read()

        clean_text = advanced_clean(text)

        save_path = os.path.join(
            OUTPUT_FOLDER,
            file
        )

        with open(
            save_path,
            "w",
            encoding="utf-8"
        ) as f:

            f.write(clean_text)

print("selesai")
```
