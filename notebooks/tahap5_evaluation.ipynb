```python
pred_df = pd.read_csv(
    f"{hasil_folder}/predictions.csv"
)

pred_df.head()
```

```python
import re

def classify_prediction(text):

    text = str(text).lower()

    if "penghinaan" in text:
        return "Penghinaan"

    elif "fitnah" in text:
        return "Fitnah"

    elif "pencemaran nama baik" in text:
        return "Pencemaran Nama Baik"

    elif "nama baik" in text:
        return "Pencemaran Nama Baik"

    else:
        return "Penghinaan"


pred_df["predicted_label"] = pred_df["predicted_solution"].apply(
    classify_prediction
)

pred_df[[
    "query",
    "actual",
    "predicted_label"
]]
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

y_true = pred_df["actual"]
y_pred = pred_df["predicted_label"]

accuracy = accuracy_score(
    y_true,
    y_pred
)

precision = precision_score(
    y_true,
    y_pred,
    average="macro",
    zero_division=0
)

recall = recall_score(
    y_true,
    y_pred,
    average="macro",
    zero_division=0
)

f1 = f1_score(
    y_true,
    y_pred,
    average="macro",
    zero_division=0
)

prediction_metrics = pd.DataFrame({
    "Metric":[
        "Accuracy",
        "Precision",
        "Recall",
        "F1"
    ],
    "Value":[
        accuracy,
        precision,
        recall,
        f1
    ]
})

prediction_metrics
```

```python
prediction_metrics.to_csv(
    f"{eval_folder}/prediction_metrics.csv",
    index=False
)

print("prediction_metrics.csv tersimpan")
```
